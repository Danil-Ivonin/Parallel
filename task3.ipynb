```python
import matplotlib.pyplot as plt
```

```python
def process_data(text_name):
    f = open(text_name + "_data.txt")
    size = int(f.readline()[:-1])
    mean_sent = float(f.readline())

    words = dict()

    for i in range(size):
        f = open(text_name + "_" + str(i) + "_words.txt")
        for i in range(10):
            data = f.readline()[:-1]
            data = data.split(':')
            if (data[0] in words):
                words[data[0]] = words[data[0]] + int(data[1])
            else:
                words[data[0]] = int(data[1])
    print("Mean sentances length = ", mean_sent)
    keys = []
    values = []
    words = dict(sorted(words.items(), key=lambda item: item[1], reverse=True))
    for k in words.keys():
        keys.append(k)
    for v in words.values():
        values.append(v)

    plt.figure(figsize=(20, 5))
    plt.bar(keys, values)
```

```python
text_name = "shinel"
process_data(text_name)
```

```python
text_name = "vishnevyi-sad"
process_data(text_name)
```
